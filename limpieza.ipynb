Aplico One-hot

```python

# Aplicar la codificación one-hot a todas las columnas categóricas
df = pd.get_dummies(df, columns=['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction'], drop_first=True)

# Mostrar las primeras filas del DataFrame final para verificar el resultado
print(df.head())
```

Hago la división de la columna "Class"

Convierto los booleanos a enteros

```python
# Convertir columnas booleanas a enteros
df = df.apply(lambda x: x.astype(int) if x.dtype == 'bool' else x)
```

```python
# Guardar el DataFrame limpio en la carpeta Downloads
ruta_descargas = r'C:\Users\Administrator\Downloads\airline_passenger_satisfaction_pruebaconisatransformados.csv'
df.to_csv(ruta_descargas, index=False)

print(f'Archivo guardado en {ruta_descargas}')
```

```python
# Supongamos que 'satisfaction' es la columna que deseas predecir
X = df.drop('satisfaction_satisfied', axis=1)  # Eliminar la columna de satisfacción de las características
y = df['satisfaction_satisfied']  # La columna de satisfacción es la variable objetivo
```

Dividir los datos de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
```

Hiperparametros

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Definir el rango de hiperparámetros a probar
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Crear el modelo base
rf = RandomForestClassifier(random_state=42)

# Crear el Grid Search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Ajustar el modelo con los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejores parámetros
best_params = grid_search.best_params_
print(f"Mejores Hiperparámetros: {best_params}")


# Obtener el mejor modelo basado en los hiperparámetros optimizados
best_rf = grid_search.best_estimator_
```

Evaluar el modelo

```python
from sklearn.metrics import accuracy_score, classification_report

# Predecir los valores del conjunto de prueba con el mejor modelo
y_pred = best_rf.predict(X_test)

# Evaluar la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión: {accuracy}")

# Mostrar un informe de clasificación detallado
print(classification_report(y_test, y_pred))
```
